```python
# df указывается для инициации функции, можно любое название, лучше чтобы они отличались от названия датасета 
# в данном приме
def prepare_square(df):
    df.loc[df['Square']>(df['Rooms']*40+20),'Square']=df['Rooms']*40+20
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    df.loc[df['Floor']>(df['HouseFloor']),'Floor']=df['HouseFloor']
    df.loc[df['Square']<(df['LifeSquare']),'Square']=df['LifeSquare']
    return df 
```

```python
# а сдесь указывается название датасета в данном примере - data
data=prepare_square(data)
```

```python
data.describe()
```

```python
data.loc[data['Rooms']==0]
```

```python
def fillna_lifeSquare(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['Square'])
    return df
```

```python
data=fillna_lifeSquare(data,data)
```

```python
data.info()
```

```python
# Создаём синтетическую переменную для того чтобы наша модель показала лучший результат
# Join_stats- синтетическая переменная
# df - некоторая переменная
#stats1 -некоторый датафрейм, пока его нет, создадим его после создания функции
#source_df-источник данных
#Join_stats1 - это датафрейм, который появится после создания функции
# mode='train'- в случае если применяем эту функцию на тренировочном датасейте, то она не нужна
#'mean_price' - столбец создавали на 5 уроке
def join_stats1(df,stats1,source_df,mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId','Square'],how='left')
    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
stats1=data.groupby(['DistrictId','Square'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
# data в конце означает источник данных так как он един указываем data
data=join_stats1(data,stats1,data)
```

```python
data.describe()
```

```python
# Разбиваем дата сет на 2 части на тренировочную и валидационную
from sklearn.model_selection import train_test_split
```

```python
features=data[['Rooms', 'Square', 'LifeSquare', 'Floor', 'mean_price']]
target=data['Price']
# Создаём датафрейм на основании тех признаков которые обработали
```

```python
X_train,X_test,y_train,y_test=train_test_split(features, target, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression as LR, Ridge
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
model1=LR()
```

```python
%%time
model1.fit(X_train,y_train)
#Обучаем модель и с помощью time засекаем время
```

```python
# Предсказываем значения
y_pred=model1.predict(X_test)
```

```python
# Определить метрику качества по валидационной выборке X_test, y_test, которая составляет 30% (test_size=0.3)
from sklearn.metrics import mean_squared_error as mse, r2_score as r2
```

```python
print(mse(y_test,y_pred),r2(y_test,y_pred))
#R2= 0.66 лучше больше
```

```python
model2=Ridge(alpha=0.05)
```

```python
%%time
model2.fit(X_train,y_train)
```

```python
# Вторая модель
y_pred_2= model2.predict(X_test)
```

```python
print(mse(y_test,y_pred_2),r2(y_test,y_pred_2))
```

```python
# Третья модель
model3=RF(n_estimators=100,max_depth=10,min_samples_leaf=5)
```

```python
%%time
model3.fit(X_train,y_train)
# Обучаем "метод случайного леса"
```

```python
y_pred_3= model3.predict(X_test)
```

```python
print(mse(y_test,y_pred_3),r2(y_test,y_pred_3))
```
